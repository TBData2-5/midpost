```
# from tensorflow.keras.applications import VGG16
from tensorflow.keras.applications import VGG16
from tensorflow.keras.layers import Input

# # load in Resnet
base = VGG16(
     # not include fully connected layers
     include_top=False,
     # weights come from imagenet
     weights="imagenet",
     # input shape
     input_shape=(224,224,3)
 )

#model summary
# base.summary()
```

```
from tensorflow.keras.layers import AveragePooling2D
from tensorflow.keras.layers import Dropout
from tensorflow.keras.layers import Flatten
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Conv2D
from tensorflow.keras.layers import BatchNormalization
from tensorflow.keras.layers import LeakyReLU
from tensorflow.keras.models import Model


# flatten layer, 972800
head = base.output
model = Flatten()(head)
# Fully Connected Layer with 475 nodes
model = Dense(475, activation = 'relu')(model)
# # Batch Normalization
# model = BatchNormalization()(model)
model = Dropout(rate=DR)(model)
# Fully Connected Layer with 95 nodes
model = Dense(95, activation = 'relu')(model)
model = Dropout(rate=DR)(model)
# # Fully Connected Layer with 49 nodes
# model = Dense(49, activation = 'relu')(model)
# model = Dropout(rate=DR)(model)
# Fully Connected Layer with 5 nodes and softmax
predictions= Dense(6, activation='softmax')(model)
model = Model(inputs = base.input, outputs = predictions)
# model.summary()
```

```
# freeze base model weights
for layer in base.layers:
  layer.trainable = False
```

```
# call back
from keras.callbacks import ReduceLROnPlateau

callback = ReduceLROnPlateau(monitor='val_loss', factor=0.1, patience=10)
# callback = tf.keras.callbacks.EarlyStopping(monitor='loss', patience=10)
```

```
early_stop = tf.keras.callbacks.EarlyStopping(
    monitor="val_loss",
    patience=6,
    verbose=0,
    baseline=None,
    restore_best_weights=True,
)
```

```
# learning rate schedule 
# for SGD
from tensorflow.keras.optimizers.schedules import ExponentialDecay
lr_schedule = ExponentialDecay(LR, decay_steps=10000, decay_rate=0.96, staircase=True)
```

```
# from tensorflow.keras.optimizers import SGD
# # compile model
# # SGD
# model.compile(SGD(learning_rate=lr_schedule), loss='categorical_crossentropy', metrics=['accuracy'])
```

```
from tensorflow.keras.optimizers import Adam
# compile model
# Adam
model.compile(Adam(lr=LR, beta_1=MO), loss='categorical_crossentropy', metrics=['accuracy'])
```

```
# plot_training function
import matplotlib.pyplot as plt
def plot_training(history, name):
  N = EPOCHS
  plt.style.use("default")
  plt.style.use("seaborn-poster")
  plt.style.use("seaborn-ticks")
  fig, ax1 = plt.subplots()
  ax2 = ax1.twinx()

  ax1.plot(np.arange(0, N), H.history["accuracy"], label="Train Accuracy", color='c', linestyle='dashed', marker='o', markersize=8)
  ax1.plot(np.arange(0, N), H.history["val_accuracy"], label="Validation Accuracy", color='r', linestyle='dashed', marker='o', markersize=8)
  ax2.plot(np.arange(0, N), H.history["loss"], label="Train Loss")
  ax2.plot(np.arange(0, N), H.history["val_loss"], label="Validation Loss")
  plt.title("Loss and Accuracy Plot {a}".format(a = name))
  ymax = max(H.history["val_accuracy"])
  xmax = np.argmax(H.history["val_accuracy"])
  text = "epoch={:.0f}, val_acc={:.4f}".format(xmax, ymax)
  bbox_props = dict(boxstyle="square,pad=0.4", fc="w", ec="k", lw=0.8)
  arrowprops = dict(arrowstyle="->",connectionstyle="angle,angleA=0,angleB=120")
  kw = dict(xycoords='data',textcoords="axes fraction", arrowprops=arrowprops, bbox=bbox_props, ha="right", va="top")
  ax1.annotate(text, xy=(xmax, ymax), xytext=(0.3,0.96), **kw)

  ax1.set_xlabel("Epoch")
  ax1.set_ylabel('Accuracy')
  ax2.set_ylabel('Loss')
  ax1.legend(bbox_to_anchor=(1.08, 1.0), loc='upper left')
  ax2.legend(bbox_to_anchor=(1.08, 0.8), loc='upper left')
  plt.savefig("midterm_plot_{a}.png".format(a = name))
  plt.show()
```

## Baseline Model: Logistic Regression

```
from sklearn.linear_model import LogisticRegression
```

```
import glob
train_path=glob.glob(r'/content/drive/MyDrive/Cassava Leaf/data/train_images/*.jpg')
train_path.sort()
print(len(train_path))
```

```
df_train['path'] = train_path
```

Since our RAM is very limited, we take a subsample from the train_images to train our logistic regression model. Since the most common class_name accounts for 61% of the train data, we want to sample 0.61*21397=13052 images from the dataset. 

```
from tqdm.notebook import tqdm 
np.random.seed(42) 
sub_train = pd.DataFrame()
sub_train = sub_train.append(df_train.sample(2000), ignore_index=True)
#Let's take a quick look at the subset
sub_train.groupby(by='class_name').count()
```

```
from sklearn.utils import shuffle
sub_train = shuffle(sub_train).reset_index(drop=True)
```

```
#Then we split the data into train data and valid data with the ratio of 8:2
from sklearn.model_selection import train_test_split

X = sub_train.drop(columns=['label'])
y = sub_train.label
X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.2, stratify=y)
```

```
#Next we resize the images and the arrays representing the images
compressed_size = (200,150)
```

```
train_array = np.array([np.asarray(Image.open(path).resize(compressed_size, Image.ANTIALIAS)) for path in tqdm(X_train.path)])
valid_array = np.array([np.asarray(Image.open(path).resize(compressed_size, Image.ANTIALIAS)) for path in tqdm(X_valid.path)])
train_array.resize(len(train_array), train_array.shape[1]*train_array.shape[2]*train_array.shape[3])
valid_array.resize(len(valid_array), valid_array.shape[1]*valid_array.shape[2]*valid_array.shape[3])
```

```
#Then we use a simple logistic regression to model
lr = LogisticRegression(class_weight='balanced', verbose=5, n_jobs=-1)
```

```
lr.fit(train_array, y_train)
```

```
preds = lr.predict(valid_array)
preds
```

```
from sklearn.metrics import confusion_matrix, classification_report, f1_score
import seaborn as sns

label = sorted(y_valid.unique())
sns.heatmap(confusion_matrix(y_valid, preds), annot=True, square=True, fmt='g', 
            xticklabels=label, yticklabels=label, cbar=False)

plt.title('Confusion matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.savefig('confusion_m.jpg')
plt.show()

print(classification_report(y_valid, preds))
```
